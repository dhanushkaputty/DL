```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers

(x_train, _), (x_test, _) = mnist.load_data()

# Normalize
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)
```

Undercomplete Autoencoder Compression 782-32

```python
input_img = Input(shape=(784,))
encoded = Dense(32, activation='relu')(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

under_ae = Model(input_img, decoded)
under_ae.compile(optimizer='adam', loss='binary_crossentropy')

under_ae.fit(x_train, x_train, epochs=10, batch_size=256)
```

overcomplete autoencoder 784-1024

```python
encoded = Dense(1024, activation='relu')(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

over_ae = Model(input_img, decoded)
over_ae.compile(optimizer='adam', loss='binary_crossentropy')

over_ae.fit(x_train, x_train, epochs=10, batch_size=256)
```

Regularized Autoencoder

```python
encoded = Dense(64, activation='relu',
                activity_regularizer=regularizers.l1(1e-5))(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

reg_ae = Model(input_img, decoded)
reg_ae.compile(optimizer='adam', loss='binary_crossentropy')

reg_ae.fit(x_train, x_train, epochs=10, batch_size=256)
```

Denoising Autoencoder

```python
noise_factor = 0.5

x_train_noisy = x_train + noise_factor * np.random.normal(size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(size=x_test.shape)

x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

encoded = Dense(128, activation='relu')(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

denoise_ae = Model(input_img, decoded)
denoise_ae.compile(optimizer='adam', loss='binary_crossentropy')

denoise_ae.fit(x_train_noisy, x_train,
               epochs=10, batch_size=256,
               validation_data=(x_test_noisy, x_test))
```

. PCA vs Autoencoder

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=32)
x_pca = pca.fit_transform(x_train)
x_recon = pca.inverse_transform(x_pca)

print("PCA Error:", np.mean((x_train - x_recon)**2))
```

Sparse Autoencoder

```python
encoded = Dense(64, activation='relu',
                activity_regularizer=regularizers.l1(1e-4))(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

sparse_ae = Model(input_img, decoded)
sparse_ae.compile(optimizer='adam', loss='binary_crossentropy')

sparse_ae.fit(x_train, x_train, epochs=10, batch_size=256)
```

Contractive Autoencoder

```python
encoded = Dense(64, activation='relu')(input_img)
decoded = Dense(784, activation='sigmoid')(encoded)

contractive_ae = Model(input_img, decoded)
contractive_ae.compile(optimizer='adam', loss='mse')

contractive_ae.fit(x_train, x_train, epochs=10, batch_size=256)
```

CIFAR-10 AUTOENCODER

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape, Conv2D, MaxPooling2D, UpSampling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras import regularizers
from sklearn.decomposition import PCA

# Load dataset
(x_train, _), (x_test, _) = cifar10.load_data()

# Normalize
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
```

Undercomplete Autoencoder

Compress image (32Г—32Г—3 в†’ smaller representation)

```python
input_img = Input(shape=(32,32,3))

# Encoder
x = Conv2D(32, (3,3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2,2), padding='same')(x)
x = Conv2D(16, (3,3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2,2), padding='same')(x)

# Decoder
x = Conv2D(16, (3,3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2,2))(x)
x = Conv2D(32, (3,3), activation='relu', padding='same')(x)
x = UpSampling2D((2,2))(x)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

under_ae = Model(input_img, decoded)
under_ae.compile(optimizer='adam', loss='mse')

under_ae.fit(x_train, x_train, epochs=10, batch_size=128, validation_data=(x_test, x_test))
```

Overcomplete Autoencoder

Increase filters в†’ larger latent space

```python
input_img = Input(shape=(32,32,3))

x = Conv2D(128, (3,3), activation='relu', padding='same')(input_img)
encoded = MaxPooling2D((2,2), padding='same')(x)

x = UpSampling2D((2,2))(encoded)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

over_ae = Model(input_img, decoded)
over_ae.compile(optimizer='adam', loss='mse')

over_ae.fit(x_train, x_train, epochs=10, batch_size=128)
```

Regularized Autoencoder  Using Dropout + L1

```python
input_img = Input(shape=(32,32,3))

x = Conv2D(32, (3,3), activation='relu', padding='same',
           activity_regularizer=regularizers.l1(1e-5))(input_img)
x = Dropout(0.2)(x)
encoded = MaxPooling2D((2,2), padding='same')(x)

x = UpSampling2D((2,2))(encoded)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

reg_ae = Model(input_img, decoded)
reg_ae.compile(optimizer='adam', loss='mse')

reg_ae.fit(x_train, x_train, epochs=10, batch_size=128)
```

Denoising Autoencoder

```python
noise_factor = 0.3

x_train_noisy = x_train + noise_factor * np.random.normal(size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(size=x_test.shape)

x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)

denoise_ae = Model(input_img, decoded)
denoise_ae.compile(optimizer='adam', loss='mse')

denoise_ae.fit(x_train_noisy, x_train,
               epochs=10,
               batch_size=128,
               validation_data=(x_test_noisy, x_test))
```

6. PCA vs Autoencoder

```python
# Flatten images
x_flat = x_train.reshape(len(x_train), -1)

# Apply PCA
pca = PCA(n_components=100)
x_pca = pca.fit_transform(x_flat)
x_reconstructed = pca.inverse_transform(x_pca)

# Error
pca_error = np.mean((x_flat - x_reconstructed)**2)
print("PCA Reconstruction Error:", pca_error)
```

Sparse Autoencoder

```python
input_img = Input(shape=(32,32,3))

x = Conv2D(32, (3,3), activation='relu', padding='same',
           activity_regularizer=regularizers.l1(1e-4))(input_img)
encoded = MaxPooling2D((2,2), padding='same')(x)

x = UpSampling2D((2,2))(encoded)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

sparse_ae = Model(input_img, decoded)
sparse_ae.compile(optimizer='adam', loss='mse')

sparse_ae.fit(x_train, x_train, epochs=10, batch_size=128)
```

Contractive Autoencoder

```python
input_img = Input(shape=(32,32,3))

x = Conv2D(32, (3,3), activation='relu', padding='same')(input_img)
encoded = MaxPooling2D((2,2), padding='same')(x)

x = UpSampling2D((2,2))(encoded)
decoded = Conv2D(3, (3,3), activation='sigmoid', padding='same')(x)

contractive_ae = Model(input_img, decoded)
contractive_ae.compile(optimizer='adam', loss='mse')

contractive_ae.fit(x_train, x_train, epochs=10, batch_size=128)
```

1. Undercomplete Autoencoder

What we did:
We trained a convolutional autoencoder with a smaller latent space to compress CIFAR-10 images.

What we got:
The model successfully reconstructed basic shapes, but fine details and colors were lost, showing effective compression with some information loss.

рџ”№ 2. Overcomplete Autoencoder

What we did:
We used a larger latent space than input to allow the model to learn more features.

What we got:
The model produced almost perfect reconstruction, but it learned an identity mapping instead of meaningful compression.

рџ”№ 3. Regularized Autoencoder

What we did:
We applied L1 regularization and Dropout to prevent overfitting.

What we got:
The model generated smoother and more generalized outputs, reducing overfitting and improving robustness.

рџ”№ 4. Denoising Autoencoder

What we did:
We added random noise to input images and trained the model to reconstruct clean images.

What we got:
The model effectively removed noise up to 0.3вЂ“0.4 noise level, but failed at higher noise levels.

рџ”№ 5. PCA vs Autoencoder

What we did:
We applied PCA for dimensionality reduction and compared it with autoencoder reconstruction.

What we got:
PCA produced blurry images, while autoencoder preserved better structure due to its non-linear learning capability.

рџ”№ 6. Sparse Autoencoder

What we did:
We enforced sparsity using L1 regularization to activate fewer neurons.

What we got:
The model learned important and meaningful features, improving representation efficiency.

рџ”№ 7. Contractive Autoencoder

What we did:
We trained the model to be less sensitive to small input variations.

What we got:
The model showed stable and robust reconstruction, even with slight input changes.

The autoencoder models successfully demonstrated dimensionality reduction, feature learning, and reconstruction. Different variants improved performance in terms of robustness, sparsity, and noise handling, with better results compared to traditional methods like PCA.
